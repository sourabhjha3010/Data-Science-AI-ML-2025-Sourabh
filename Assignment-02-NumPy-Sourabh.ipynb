## Problem: 6

Create an array $x$ with 100 points in the interval $[-1,1]$ (the points must be uniformly spaced). Then create another array $y$. 

Where $y = x^2 sin(x^2) + x$. 

Try to plot $y$ vs $x$ using matplotlib.

```python
# Write your code here
import matplotlib.pyplot as plt
x = np.linspace(-1, 1, 101)
y = (x ** 2) * (np.sin(x ** 2)) + x

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("Plot x Vs y")
plt.grid(True)
plt.show()
```

## Problem: 7 
Create a function which creates an $n√ón$ array with $(i , j)$ -entry equal to $i+ j$

```python
# Write your code here
n = int(input("Enter the value of n to create nXn array: "))
def array(n):
    arr =np.zeros((n,n),dtype='int')
    for i in range(n):
        for j in range(n):
            arr[i,j] =i+j

    return arr
print(f" nXn array created as: \n {array(n)}")
```

# Problem: 8
Evaluate $cos$ and $sin$ on the interval $[0,1]$  with 1000 points and then stack the results into a tall array with rows being the $(cos(x),sin(x))$ entries.

```python
# Write your code here
x = np.linspace(0,1,1000)
sinx = np.sin(x)
cosx = np.cos(x)
y = np.stack((cosx, sinx), axis=1)
print(y)
plt.plot(sinx,cosx)
plt.xlabel('sinx')
plt.ylabel('cosx')
plt.title("Plot sinx Vs cosx")
plt.grid(True)
plt.show()
```
